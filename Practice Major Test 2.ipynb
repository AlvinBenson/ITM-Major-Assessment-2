### Basic Queries

What are the top three (3) remittance months for the United States?

Do not forget to remove 2018 and 2019 columns.

Work with copies (via .copy()) instead of assigning the original dataframe

Hints:
* Use Pandas Dataframe `transpose()`
* Don't forget to set inplace as needed

Sources:  
Drop rows by index:
https://stackoverflow.com/questions/14661701/how-to-drop-a-list-of-rows-from-pandas-dataframe

```python
unwanted_cols_first = [2018,2019]
df_tempo = df_country[df_country["Country"].isin(["U.S.A."])].copy()
df_tempo.drop(columns=unwanted_cols_first,inplace=True)
df_tranposed = df_tempo.transpose().copy()

```

```python
df_tranposed.drop("Country",axis=0,inplace=True)
df_tranposed.rename(columns={66:"remittances"},inplace=True)
df_tranposed = df_tranposed.apply(pd.to_numeric)
df_tranposed.nlargest(3,"remittances")
```

```python
df_remittances = df_tranposed.reset_index()
df_remittances[["remittances","index"]].sort_values("remittances",ascending=False)
```

### Data Visualization

Specifications:
* Figure consists of 1 row by 2 columns
* Figure size is 12 inches x 20 inches
* Amounts must be in thousands (so you'll have to divide by 1000)

Useful links:  
https://stackoverflow.com/questions/10998621/rotate-axis-text-in-python-matplotlib
https://stackoverflow.com/questions/10998621/rotate-axis-text-in-python-matplotlib

```python
df_remittances = df_tranposed.reset_index()
df_remittances[["remittances","index"]].sort_values("remittances",ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
fig,axesbar = plt.subplots(3,1,figsize=(12,20))



axesbar[0].barh(df_country_sorted.nlargest(10,2019).sort_values(by=2019)["Country"],df_country_sorted.nlargest(10,2019).sort_values(by=2019)[2019]/1000,color="Green")
axesbar[0].set_title("Top 10 Sources of Cash Remittances")
axesbar[0].set_xlabel("Remittances (in Thousands)")
axesbar[0].set_ylabel("Country")

axesbar[1].bar(df_continent_sorted["Country"],df_continent_sorted[2019]/1000,color="Orange")
axesbar[1].set_title("Top 10 Continents for Cash Remittances")
axesbar[1].set_xlabel("Continents")
axesbar[1].set_ylabel("Remittances (in Thousands)")

axesbar[2].bar(df_remittances[["remittances","index"]].sort_values("remittances",ascending=False)["index"],df_remittances[["remittances","index"]].sort_values("remittances",ascending=False)["remittances"]/1000,color="Red")
axesbar[2].set_title("Monthly Cash Remittances")
axesbar[2].set_xlabel("Month")
axesbar[2].set_ylabel("Remittances (in Thousands)")

plt.show()
```
